```python

import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("C:\\Users\\supri\\Downloads\\train.csv")


df = df.dropna()
df = df[df["fare"] > 0]
```

```python
baseline_features = [
    "trip_duration",
    "distance_traveled",
]

X_base = df[baseline_features]
y = df["total_fare"]
```

```python


Xb_train, Xb_test, y_train, y_test = train_test_split(
    X_base, y, test_size=0.2, random_state=42
)

```

```python

# ---------------- BASELINE ----------------
baseline_model = tf.keras.Sequential([
    layers.Dense(64, activation="relu", input_shape=(Xb_train.shape[1],)),
    layers.Dense(32, activation="relu"),
    layers.Dense(1)
])

baseline_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

baseline_model.fit(Xb_train, y_train, epochs=10, batch_size=32)
baseline_mae = baseline_model.evaluate(Xb_test, y_test)[1]
```

```python
#feature engineering
df["distance_km"] = df["distance_traveled"]
df["num_of_passengers"] = df["num_of_passengers"].clip(1, 6)

engineered_features = [
    "trip_duration",
    "distance_km",
    "num_of_passengers",
    "surge_applied"
]

```

```python
Xe = df[engineered_features]

scaler = StandardScaler()
Xe_scaled = scaler.fit_transform(Xe)
Xe_train, Xe_test, ye_train, ye_test = train_test_split(
    Xe_scaled, y, test_size=0.2, random_state=42
)
```

```python
fe_model = tf.keras.Sequential([
    layers.Dense(64, activation="relu", input_shape=(Xe_train.shape[1],)),
    layers.Dense(32, activation="relu"),
    layers.Dense(1)
])

fe_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

fe_model.fit(Xe_train, ye_train, epochs=10, batch_size=32)
fe_mae = fe_model.evaluate(Xe_test, ye_test)[1]

print("Baseline MAE:", baseline_mae)
print("Feature Engineered MAE:", fe_mae)






```
